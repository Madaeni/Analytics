Получить матрицу At путем транспонирования матрицы A.

```python
At = A.T
At
```

```python
B = np.dot(At, A)
B
```

Какой размер имеет матрица B?

```python
B.size
```

Получится ли вычислить обратную матрицу для матрицы B и почему?

Зная, что квадратная матрица обратима тогда и только тогда, когда она невырождена, то есть её определитель не равен нулю. Для неквадратных матриц и вырожденных матриц обратных матриц не существует.

```python
#Вычислим определитель матрицы B:
np.linalg.det(B)
```

Определитель равен нулю, следовательно обратной матрицы не существует.

###### 13. Приравняйте к нулю отрицательные числа в матрице D_inv, а положительные - к единице. Убедитесь, что в матрице D_inv остались только нули и единицы.

Элементы матрицы E, для которых соответствующий элемент матрицы D_inv равен 1, должны быть равны соответствующему элементу матрицы B, а элементы матрицы E, для которых соответствующий элемент матрицы D_inv равен 0, должны быть равны соответствующему элементу матрицы C.

```python
E = np.where(D_inv > 0, B, C)
E
```
